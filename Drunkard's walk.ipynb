```python
import numpy as np
import matplotlib.pyplot as plt 

#位置从0开始
position = 0

#记录每一步走的距离
walk = [position]

#设置走1000步
steps = 1000

for i in range(steps):
    #如果随机数为0，step=1，向前走一步
    #如果随机数为1，step=-1，向后走一步
    step = 1 if np.random.randint(0,2) else -1
    
    #现在的位置加上step，向前走一步或者向后走一步
    position = position + step
    
    #记录位置
    walk.append(position)

    #画图
pic = plt.plot(walk)
```
